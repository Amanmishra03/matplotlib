#  Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

c = ['r','g',"y","b",'r','y','b']
sizes = [100,200,150,90,80,110,250]

plt.scatter(day,no, color=c, s=sizes, alpha=0.4,marker='*', edgecolor='r',linewidths=2)

plt.xlabel('day', fontsize=15)
plt.ylabel('No.', fontsize=15)
plt.title("Scatter plot", fontsize=15)

plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

colors = [10,49,52,8,66,77,12]
sizes = [100,200,150,90,80,110,250]

plt.scatter(day,no, c=colors, s=sizes)

plt.xlabel('day', fontsize=15)
plt.ylabel('No.', fontsize=15)
plt.title("Scatter plot", fontsize=15)

plt.show()
```
